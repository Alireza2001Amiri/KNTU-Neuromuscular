```python
scaler = MinMaxScaler()
dfin = pd.DataFrame(scaler.fit_transform(dfi), columns=dfi.columns)
```

# define value of K in clustering

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
```

```python
k_list = range(2, 11)
sse = []
sil = []

# Calculate SSE and silhouette score for different k
for k in k_list:
    km = KMeans(n_clusters=k)
    km.fit(dfin)
    sse.append(km.inertia_)
    sil.append(silhouette_score(dfin, km.labels_))
```

```python
# Plot SSE and silhouette score against k
plt.figure(figsize=(16,6))
plt.subplot(1, 2, 1)
plt.plot(k_list, sse, '-o')
plt.xlabel('Number of clusters k')
plt.ylabel('Sum of squared distances')
plt.title('Elbow Method')

plt.subplot(1, 2, 2)
plt.plot(k_list, sil, '-o')
plt.xlabel('Number of clusters k')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Method')
plt.show()

```
